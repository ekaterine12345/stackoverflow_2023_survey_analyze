## Ploting a bar chart for categories of age

```python
age_counts = df['Age'].value_counts(normalize=True)

age_categories = age_counts.index.tolist() # store categoris of age in a list
percents = (age_counts.to_numpy() * 100).round(2) # store percantage of each category in a list
```

```python
fig, ax = plt.subplots(figsize =(8, 4)) # determine the size of the figure
plt.title(schema_df.loc['Age', 'question'][:-2], fontsize=20) # from questions' dataframe fetching a question about the age column

ax.barh(age_categories, percents, color="darkblue") # plot the horizontal bar chart

for bor in ['top', 'bottom', 'left', 'right']: # Removing all borders
    ax.spines[bor].set_visible(False)

ax.xaxis.set_ticks_position('none') #  removing X and Y ticks
ax.yaxis.set_ticks_position('none')

ax.xaxis.set_tick_params(pad = 8) # Adding a padding
ax.yaxis.set_tick_params(pad = 7)

ax.invert_yaxis() # Sort by descending order of percentages

for i in ax.patches: # On each bar writing the appropriate percantage with % sign
    plt.text(i.get_width()+0.3, i.get_y()+0.5, str(round((i.get_width()), 2))+"%",
             fontsize = 9, fontweight ='bold',
             color ='black')
plt.show()
```

## ploting a pie chart based on AI usage

Calculating the number of developers who are using AI tools;
how many of them are going to use AI tools and how many of them
are not going to use AI tools.
Based on this 3 categories of answers I will plot the pie chart

```python
AISel_count = df['AISelect'].value_counts(normalize=True)
AISelect_labels = AISel_count.index.tolist()
AISel_percents = (AISel_count.to_numpy() * 100).round(2)
```

```python
plt.title(schema_df.loc['AISelect', 'question'][:-2], fontsize=20) # writing appropriate question as title
plt.pie(AISel_percents, labels=AISelect_labels, shadow=True, colors=['darkgreen', 'pink', 'yellow']) # plot the pie chart
plt.legend(title="AI Usage",  bbox_to_anchor=(1.3,0.25))  # creating legend and positioning it at the left bottom 
```

## Percentage of usage for each online courses platformas

There are selected popular sites that are offering courses.
I am calculating the percantage for each of them to see how many respodents are using that site

```python
courses_list = ['Udemy', 'Coursera', 'Codecademy', 'Pluralsight', 'edX', 'Udacity', 'Skillsoft'] 
courses_percent = []
for each in courses_list:
    courses_percent.append((df['LearnCodeCoursesCert'].str.contains(each).value_counts(normalize=True)[True]*100).round(2)) 
courses_percent = np.array(courses_percent)
```

```python
fig, ax = plt.subplots(figsize =(16, 6))
plt.title("Percentage of developers using online courses platformas", fontsize=20) # dispaying a title
ax.bar(courses_list, courses_percent, color="orange", edgecolor="grey", linewidth=2) # ploting a bar chart

for bor in ['top', 'bottom', 'left', 'right']: # 
    ax.spines[bor].set_visible(False) # Removing all borders

ax.xaxis.set_ticks_position('none') # Removing X ticks

for i in ax.patches: # for each bar writing its percantage
    plt.text(i.get_x()+0.25, i.get_height()+0.5, str(round((i.get_height()), 2))+"%",
             fontsize = 9, fontweight ='bold',
             color ='black')
```
